```python
class cafe:
    
    def __init__(self, name, city, employees, tables):
        self.name = name
        self.city = city
        self.employees = employees
        self.tables = tables
        self.menu = dict()
        self.product_sold = dict()

    def rev_change(self, percentage):
        """This function will add an employee if the revenue has increased by 50% or more
        and will remove one employee if the revenue has decreased by 50% or less"""
        if percentage>=0.5:
            self.employees+=1
        if percentage <= -0.5:
            self.employees-=1
            
    def revenue(self):
        """This function will calculate the revenue generated by the cafe by each product"""
        print({i:self.menu[i]*self.product_sold[i] for i in self.menu})
    
    def menu_update(self, increase):
        """This function will increase the price of the most selling product by the percentage given as input"""
        import operator
        max = sorted(self.product_sold.items(), key =operator.itemgetter(1))[-1][0]
        self.menu[max]=self.menu[max]*(1+increase)

    
```

```python
central_perk = cafe('Central Perk', 'NYC', 4, 20)
```

```python
central_perk.rev_change(0.6)
```

```python
central_perk.employees
```

```python
central_perk.rev_change(0.4)
central_perk.employees
```

```python
central_perk.rev_change(-0.5)
central_perk.employees
```

```python
central_perk.menu = {'Regular coffee' : 1.99, 'Macchiato' : 3.99, 'Capuccino' : 2.99}
central_perk.product_sold = {'Regular coffee' : 100, 'Macchiato' : 50, 'Capuccino' : 75}
```

```python
central_perk.revenue()
```

```python
central_perk.menu_update(0.1)
```

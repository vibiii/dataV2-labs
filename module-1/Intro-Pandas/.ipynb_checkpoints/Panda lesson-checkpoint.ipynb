```python
house_dict = {'Baker House': [7420, 'BrkSide', 1939, 5, 118000],
              'Beazley House': [14115, 'Mitchel', 1993, 5, 143000],
              'Dominguez House': [14260, 'NoRidge', 2000, 8, 250000],
              'Hamilton House': [6120, 'OldTown', 1931, 7, 129900],
              'James House': [11250, 'CollgCr', 2001, 7, 223500],
              'Martinez House': [9600, 'Veenker', 1976, 6, 181500],
              'Roberts House': [9550, 'Crawfor', 1915, 7, 140000],
              'Smith House': [8450, 'CollgCr', 2003, 7, 208500],
              'Snyder House': [10084, 'Somerst', 2004, 8, 307000],
              'Zuckerman House': [10382, 'NWAmes', 1973, 7, 200000]}
```

```python
pd.DataFrame(house_dict)
```

```python
# Transpose
pd.DataFrame(house_dict).transpose()
```

```python
# Transpose and rename the columns
colnames =['LotSize','Neighborhood','YearBuilt','Quality','SalePrice']
df= pd.DataFrame(house_dict).transpose()
df.columns = colnames
df
```

```python
# Retrieve the built year of Smith House
df['YearBuilt'] ['Smith House']
```

```python
# Fletch all the raw where LotSize > 10000
"""Dataframe for DataFrame Lotsize > 10000"""
df[df['LotSize']>10000]
```

```python
# Do the save and lock the dataframe so that later if we want to make a change on this sub-dataframe, it will be taken into account on the main dataframe

df.loc[df['LotSize']>10000]
```

```python
# Fletch all the raw where LotSize > 1000 and Quality >5
df[(df.LotSize>1000) & (df.Quality>5)]
```

```python
# Fletch all the raw where LotSize > 1000 and Quality >5 and show only SalePrice and Quality columns
df[(df.LotSize>1000) & (df.Quality>5)][['SalePrice','Quality']]
```

```python
# Show only the columns from 0 to 3 and the raws from 1 to 4 
df.iloc[0:3,1:4]
```
